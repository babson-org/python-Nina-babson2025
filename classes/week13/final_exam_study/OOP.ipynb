Basic OOP1 Practice Code

```python
class Dog:
    def __init__(self, name, age, breed):
        self.name = name
        self.age = age
        self.breed = breed

mydog = Dog("Buddy", 12, "poodle")
print(mydog.name)
print(mydog.age)

```

OOP 2 Practice Code

```python
class Car:
    def __init__(self, make, model, year, speed=0):
        self.make = make
        self.model = model
        self.year = year
        self.speed = speed

    def __str__(self):
        return f"{self.year} {self.make} {self.model} traveling at {self.speed} mph"

print(Car("Tesla", "Model S", 2024, 70))
```

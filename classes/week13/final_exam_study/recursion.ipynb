```python
#Easy Example 1
def countdown(n):
    if n == 0:
        print("Program Complete!")
    else:
        print(n)
        countdown(n-1)

countdown(5)
```

Map()
:transform every element in a list/apply this function to every item

```python
nums = [1,2,3,4,5]
result = list(map(lambda x: x*2, nums))
print(result)
```

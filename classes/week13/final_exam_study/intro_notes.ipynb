Loops

```python
### While Loops ###
x = 0
while x < 3:
    print(x)
    x += 1
# output 0 1 2

x = 0
while x < 3:
    x += 1
    print(x)
#output 1 2 3

### For Loops ###
for i in range(3):
    print(i)
```

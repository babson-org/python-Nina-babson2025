Loops

```python
### While Loops ###
x = 0
while x < 3:
    print(x)
    x += 1
# output 0 1 2

x = 0
while x < 3:
    x += 1
    print(x)
#output 1 2 3

### For Loops ###
for i in range(3):
    print(i)
```

Built in Functions

Math

```python
myList = [9,1,8,2,7,3,6,4,5]

print(max(myList))

print(min(myList))

print(sum(myList))
```

Iterable

```python
myList = [9,1,8,2,7,3,6,4,5]

len(myList)

sorted(myList)

list(reversed([1,2,3]))
list(reversed(myList))
```

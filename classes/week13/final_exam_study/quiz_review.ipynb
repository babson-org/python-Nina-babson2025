```python
### Diamond Printing ###
# fixed size of diamond - determine how many lines you want diamond to be *must be odd number

#hard coding dimensions to avoid user error

# Top half of diamond
n = 5  # height of the top half
for i in range(n):
    print(" " * (n - i - 1) + "*" * (2*i + 1))

# Bottom half of diamond
for i in range(n - 2, -1, -1):
    print(" " * (n - i - 1) + "*" * (2*i + 1))
```
